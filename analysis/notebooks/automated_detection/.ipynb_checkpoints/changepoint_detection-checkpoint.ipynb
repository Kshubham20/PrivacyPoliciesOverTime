```python
kpss_index = ["kpss_stat","kpss_p_value"]
def do_kpss(s):
    kpss_stat,p_value,lags,crit = tsatools.kpss(s)
    return pd.Series([kpss_stat,p_value],index=kpss_index)

bds_index = ["bds_stat","bds_p_value"]
def do_bds(s):
    bds_stat,p_value = tsatools.bds(s)
    return pd.Series([bds_stat,p_value],index=bds_index)

adf_index = ["adf_stat","adf_p_value"]
def do_adf(s):
    try:
        adf_stat,p_value,*_ = tsatools.adfuller(s)
    except:
        return None
    return pd.Series([adf_stat,p_value],index=adf_index)

acv_index = ["acv_mean","acv_median","acv_max","acv_min"]
def do_acv(s):
    acv = tsatools.acovf(s)
    return pd.Series([np.mean(acv),np.median(acv),np.max(acv),np.min(acv)],index=acv_index)

#FIXME
df_alexa_common = df_alexa[(df_alexa >= THRESH).any(axis=1)]

df_kpss = df_alexa_common.progress_apply(do_kpss,axis=1)
df_kpss = df_kpss[df_kpss['kpss_stat'].notna()]
df_kpss["kpss_p_value_adj"] = multipletests(df_kpss["kpss_p_value"])[1]

df_bds = df_alexa_common.progress_apply(do_bds,axis=1)
df_bds = df_bds[df_bds['bds_stat'].notna()]
df_bds["bds_p_value_adj"] = multipletests(df_bds["bds_p_value"])[1]

err_settings = np.seterr(all="raise")
try:
    df_adf = df_alexa_common.progress_apply(do_adf,axis=1)
finally:
     np.seterr(**err_settings)
df_adf = df_adf[df_adf['adf_stat'].notna()]
df_adf["adf_p_value_adj"] = multipletests(df_adf["adf_p_value"])[1]

err_settings = np.seterr(all="raise")
try:
    df_acv = df_alexa_common.progress_apply(do_acv,axis=1)
finally:
     np.seterr(**err_settings)
df_acv = df_acv[df_acv['acv_mean'].notna()]
#df_za["acv_p_value_adj"] = multipletests(df_za["f_p_value"])[1]

df_scores = pd.concat([df_kpss,df_bds,df_adf,df_acv],axis=1).sort_values("kpss_p_value")
```

```python
def plot_top(df,rows):
    
    print(list(rows))
    
    plt.figure(figsize=(12,5))
    
    labels=list(df)
    x=list(range(len(labels)))

    ys=[(df.loc[rowId],rowId) for rowId in rows.index]

    #Plot
    for y,rowId in ys:
        fig = sns.lineplot(x=labels,y=y,label=rowId)
        
    plt.ylim(bottom=0)
        
    for item in fig.get_xticklabels():
        item.set_rotation(45)
        
    plt.legend(loc='upper left', bbox_to_anchor=(0,-0.15))
    plt.ylabel("Document frequency")
    plt.title(rows.name)
        
    plt.show()

def plot_scores(df,scores,key=None):
    
    scores=scores.sort_values(key=key)
    
    first = scores.iloc[:10]
    last = scores.iloc[-10:]
    
    plot_top(df,first)
    plot_top(df,last)
```

```python
scorefs = bds_index + kpss_index + adf_index + acv_index
for scoref in scorefs:
    plot_scores(df_alexa,df_scores[scoref])
```

# Changepoints

```python
def plot_changepoint_count(df,changepoint_df):
    fig = plt.figure(figsize=(7,3))
    

    ncolors=len(changepoint_df.ngrams.unique())
    
    #Plot
    fig = sns.lineplot(x="interval",y="changes",hue="ngrams",palette=sns.color_palette("hls", ncolors),data=changepoint_df)
        
    plt.ylim(bottom=0)
        
    for item in fig.get_xticklabels():
        item.set_rotation(45)
        
    plt.legend(loc='upper left',ncol=2)
    
    plt.ylabel("Number of changes")

    plt.xlim(0,21)
    plt.xlabel("")
        
#     plt.show()
    
    return fig
```

```python
model = "l1"


changepoint_df = pd.DataFrame()

breakpoint_lookups = {}

for n in range(1,9):
    
    #Load the data
    fn = f"../../data/{n}grams.pkl"
    df_i_all = pd.read_pickle(fn)
        
    df_i_alexa = df_i_all[ys_values[24:]].apply(lambda s: s / policy_counts[s.name],axis=0)

    breakpoint_counter = collections.Counter()
    breakpoint_up_counter = collections.Counter()
    breakpoint_down_counter = collections.Counter()

    df_i_alexa_common = df_i_alexa[(df_i_alexa >= 0.01).any(axis=1)]
    cols = list(df_i_alexa_common)
    print(len(df_i_alexa_common), "/", len(df_i_alexa),'\n')

    breakpoint_lookup = collections.defaultdict(list)
    breakpoint_lookups[n] = breakpoint_lookup

    for text, signal in tqdm(df_i_alexa_common.iterrows(),total=len(df_i_alexa_common)):
        signal = signal.to_numpy()
        algo = rpt.Pelt(model=model, min_size=1, jump=1).fit(signal)
        breakpoints = algo.predict(pen=0.01)
        breakpoints = [b for b in breakpoints if b != len(cols)]

        breakpoint_counter.update(breakpoints)

        for bkp in breakpoints:
            breakpoint_lookup[cols[bkp]].append(text)
            if signal[bkp+1] > signal[bkp-1]:
                breakpoint_up_counter.update([bkp])
            else:
                breakpoint_down_counter.update([bkp])
            
    max_bkpc = max(breakpoint_counter.values())
    num_pols = len(df_i_alexa_common)
        
    for t,counter in [("all",breakpoint_counter),("up",breakpoint_up_counter),("down",breakpoint_down_counter)]:
        changepoint_df = changepoint_df.append(
            pd.DataFrame([(cols[bkp],counter[bkp],counter[bkp]/max_bkpc,counter[bkp]/num_pols,t,f"{n}-grams") for bkp in range(len(cols))],
                         columns=["interval","changes","perc_changes","perc_changes_pols","type","ngrams"]))
```

```python
fig = plot_changepoint_count(df_alexa_common,changepoint_df[(changepoint_df.type == "all") & changepoint_df.ngrams.isin(["1-grams","3-grams","8-grams"])])

fn = "../figures/changepoints.pdf"
plt.savefig(fn, bbox_inches='tight')
```

```python
def plot_topics(df_topics):
    
    plt.figure(figsize=(12,5))
    
    labels=list(df)
    x=list(range(len(labels)))

    ys=[(row,rowId) for rowId,row in df_topics.iterrows()]

    #Plot
    for y,rowId in ys:
        fig = sns.lineplot(x=labels,y=y,label=rowId)
        
    plt.ylim(bottom=0,top=1)
        
    for item in fig.get_xticklabels():
        item.set_rotation(45)
        
    
        
    plt.legend(loc='upper left', bbox_to_anchor=(0,-0.15))
    #plt.ylabel(rows.name)
        
    plt.show()

def get_top_topics(n,ys,topN=10):
    fn = f"../../data/{n}grams.pkl"
    df_n = pd.read_pickle(fn).apply(lambda s: s / policy_counts[s.name],axis=0)
    
    rowIds = breakpoint_lookups[n][ys]
    signals = df_n.loc[rowIds]
    top_signals = list(reversed(list(signals.sort_values(ys).index)))
    return df_n.loc[top_signals[:topN]]

plot_topics(get_top_topics(3,"2017_A"))
```

```python
def get_x(ys_vals,month_and_year):
    month, year = month_and_year
    
    first_ys_value = min(list(ys_vals))
    start_year = int(first_ys_value[:4])
    start_interval = first_ys_value[5]
    if start_interval == "A":
        start_month = 3
    else:
        start_month = 9
        
    x = (year - start_year) * 2 + (month - start_month) * (1/6)
    
    return x

def add_point_at(ys_vals,month_and_year,y,label,**kwargs):
    
    x = get_x(ys_vals,month_and_year)
    plt.plot(x,y,'ko')
    plt.text(x,y,label,verticalalignment="center",**kwargs)
    
def add_line_at(ys_vals,month_and_year,label,**kwargs):
    x = get_x(ys_vals,month_and_year)
    plt.axvline(x,color="black",linestyle="--")
    transform = plt.axes().get_xaxis_transform()
    plt.text(x,1,label,transform=transform,**kwargs)
    
def set_x_ticks(ys_values):
    ys_values = sorted(ys_values)
    num_intervals = len(ys_values)
    unique_years = sorted(set([int(interval.split("_")[0]) for interval in ys_values]))
    plt.xticks(np.arange(0, num_intervals, 2)-0.5, unique_years, rotation="vertical")
    
```
